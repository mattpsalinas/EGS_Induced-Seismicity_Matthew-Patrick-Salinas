```python

from scipy.interpolate import make_interp_spline
from scipy.signal import savgol_filter

# Generate some sample data
np.random.seed(0)
x = np.linspace(0, 10, 50)
y = np.sin(x) + np.random.normal(0, 0.5, size=x.size)

# Plot the original scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='gray', label='Original Data', alpha=0.6)

# Step 1: Smooth Trendline using Cubic Spline
x_smooth = np.linspace(x.min(), x.max(), 500)  # More points for a smoother line
spline = make_interp_spline(x, y, k=3)  # Cubic spline
y_smooth = spline(x_smooth)
plt.plot(x_smooth, y_smooth, color='blue', label='Cubic Spline Fit')

# Step 2: Smooth Data using Savitzky-Golay Filter
y_smoothed = savgol_filter(y, window_length=7, polyorder=2)  # Moving average-like smoothing
plt.plot(x, y_smoothed, color='red', label='Savitzky-Golay Smoothed', linestyle='--')

# Customize and show the plot
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Smoothed Scatter Plot with Trendline')
plt.legend()
plt.grid(True)
plt.show()
```
